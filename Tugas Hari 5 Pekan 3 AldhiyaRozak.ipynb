````python
from matplotlib import pyplot as plt
from scipy import stats
import seaborn as sns
import numpy as np
import pandas as pd
````

````python
df = pd.read_csv('diamonds.csv')
````

# Soal 1: Scatter Plot

Buatlah scatterplot menggunakan Seaborn dari data df di atas, variasikan menggunakan palette yang telah disediakan sehingga didapatkan output sesuai expected.

````python
fig, ax = plt.subplots(figsize=(13, 13))
palette="ch:r=-.2,d=.3_r"
#code here
sns.scatterplot(df["carat"],df["price"],hue=df["clarity"],hue_order=["I1","SI2","SI1","VS2","VS1","VVS2","VVS1","IF"],size=df["depth"],palette=palette,sizes=(1,10),linewidth=0)
plt.show()
````

````python
#code here
def hexbin(x, y, color, **kwargs):
    cmap = sns.light_palette(color, as_cmap=True)
    plt.hexbin(x, y, gridsize=15, cmap=cmap, **kwargs)
with sns.axes_style("dark"):
    fig = sns.FacetGrid(df,hue="cut",col="cut",height=4)
fig.map(hexbin,"carat","price",extent=[0,3.0,0,20000])
plt.show()
````

# Soal 3: Pair Plot

Ambil 300 data secara acak dari df, kemudian buat pair plotnya dari 300 data tersebut untuk kolom carat, depth, table, dan price sehingga dihasilkan output sesuai expected.

````python
df = df.sample(n=300, random_state=123)

#code here
def hide_current_axis(*args, **kwds):
    plt.gca().set_visible(False)
a = sns.pairplot(df,hue="cut",vars=["carat","depth","table","price"])
a.map_upper(hide_current_axis)
plt.show()
````

# Soal 4: Joint Plot

Gunakan data yang sama dengan soal 3 untuk menghasilkan dua visualisasi joint plot antara carat dan price nya sesuai expected output.

````python
#code here
with sns.axes_style("dark"):
    sns.jointplot(df["carat"],df["price"],kind="kde")
with sns.axes_style("dark"):
    sns.jointplot(df["carat"],df["price"],kind="reg",stat_func=stats.pearsonr)
````
