Dengan menggunakan data di bawah ini, plot kan data tersebut kemudian custom sehingga memiliki result seperti yang di harapkan :

note: warna terdiri dari warna Hijau, kuning, Biru, dan Merah

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 100)
cos = np.cos(x)
cos3 = np.cos(x)*3
sin = np.sin(x-90)
sin2 = np.sin(x)*2
```

```python
# code here
plt.rcParams["figure.figsize"] = (9.5,6)
plt.plot(x,cos,color="r",label="Cosx")
plt.plot(x,cos3,color="g",linestyle="--",label="3 Cosx")
plt.plot(x,sin,color="b",linestyle="-.", label="Sin(x-90)")
plt.plot(x,sin2,color="y",marker="x",linestyle="", label="2 Sinx")
plt.xlabel("X-Axis Label")
plt.ylabel("Y-Axis Label")
plt.title("Grafik Cos Sin")
plt.legend(loc="center")
plt.savefig("Grafik.png")
plt.show()
```
